```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
raw = [['total_energy_consumption', 'core_freq'], ['total_energy_consumption', 'num_cores'], ['total_energy_consumption', 'kernel.max_pids'], ['total_energy_consumption', 'vm.drop_caches'], ['total_energy_consumption', 'kernel.sched_nr_migrate'], ['total_energy_consumption', 'emc_freq']]
```

```python
left_nodes = [tup[1] for tup in raw]

G = nx.DiGraph()
G.add_edges_from(raw)

pos = nx.bipartite_layout(G, nodes=left_nodes, align='horizontal')

node_colors = ["turquoise" if node in left_nodes else "pink" for node in G.nodes]

fig = plt.figure(1, figsize=(20, 6), dpi=100)

nx.draw(G,pos=pos, with_labels=True, node_color = node_colors)

plt.show()
```
